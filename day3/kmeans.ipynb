```python
import pandas 
df = pandas.read_csv('https://raw.githubusercontent.com/ltdaovn/dataset/master/ABC_Customers.csv')
print(df)
```

```python
# data explore
X = df.iloc[:, [3,4]]
import matplotlib.pyplot as plt 
plt.scatter(X.iloc[:, 0], X.iloc[:, 1])
plt.show()
```

```python

# choose k
from sklearn.cluster import KMeans

inertias = []
for n_cluster in range(1, 11): # elbow
    model = KMeans(n_clusters=n_cluster).fit(X)
    inertias += [model.inertia_]

plt.plot(range(1,11), inertias)
plt.show()


```

```python
# using k = 4
n_cluster = 5
model = KMeans(n_clusters=n_cluster).fit(X)
Y = model.predict(X)

print('Predict values:', Y)

for i in range(n_cluster):
    plt.scatter(X.iloc[Y == i, 0], X.iloc[Y == i, 1])

plt.show()
```

```python

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)

tree.plot_tree(clf, filled=True) 
```
